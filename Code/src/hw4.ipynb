#### Bullet 5: Observation on embedding pairs

```python
from hw4_utils import load_data
import numpy
train_set, valid_set, test_set, dic = load_data(3)
embeddings = numpy.load('../result/embeddings.npy')
```

```python
dic['words2idx']
```

```python
dic_words2idx = dic['words2idx']
words2vectors = dict()
for key, value in dic_words2idx.items():
    words2vectors[key] = embeddings[value]
```

```python
words2vectors
```

###### Findings are in the comments

```python
from scipy.spatial.distance import cosine 

# I could observe several interesting patterns from the data
# First of all lets have a look if we could learn that we are looking for an airline to fly to Florida 
print(cosine(words2vectors['florida'] - words2vectors['delta'], words2vectors['florida'] - words2vectors['united']))
# prints 0.136188182621, which is low enough to confirm we could learn that we are looking for an airline for Florida

# What if I try Lufthansa instead of United from previos
print(cosine(words2vectors['florida'] - words2vectors['delta'], words2vectors['florida'] - words2vectors['lufthansa']))
# prints 0.410835894191. We see big increase between local airline and international airline.
# Perhaps people are more accustomed for local airlines, and look for local airlines for local flights.
# Thus, we had much smaller results for Delta vs United rather than Delta vs Lufthansa

# Lets try other local airlines to see if pattern persists
print(cosine(words2vectors['florida'] - words2vectors['delta'], words2vectors['florida'] - words2vectors['continental']))
# prints 0.119208534852
print(cosine(words2vectors['florida'] - words2vectors['delta'], words2vectors['florida'] - words2vectors['american']))
# prints 0.149490773751
# As we have expected Delta vs local airlines are much closer rather than Delta vs Lufthansa.
# This makes sense since, we would expect some kind of similar patterns for local airlines for local flights
# rather than international airlines for local flights

# Lets look at other results
print(cosine(words2vectors['florida'] - words2vectors['flight'], words2vectors['georgia'] - words2vectors['flight']))
# prints 0.079974350448
# We are learning that Florida and Georgia are states which you can take flights to

# Lets see if we are learning months
print(cosine(words2vectors['november'] - words2vectors['month'], words2vectors['december'] - words2vectors['month']))
# prints 0.210565320508
# Looks like we are indeed learning that November and December are months

# Lets see how breakfast and dinner related in terms of cost
print(cosine(words2vectors['dinner'] - words2vectors['cost'], words2vectors['breakfast'] - words2vectors['cost']))
# prints 0.179627447847
# We are getting some kind of relation between breakfast and dinner in terms of cost

# Lets see how cost and dollars are related
print(cosine(words2vectors['highest'] - words2vectors['cost'], words2vectors['lowest'] - words2vectors['cost']))
# prints 0.286330131367
print(cosine(words2vectors['highest'] - words2vectors['dollars'], words2vectors['lowest'] - words2vectors['dollars']))
# prints 0.230120793979
# First of all we see that highest and lowest are related in terms of cost.
# Also we are seeing that they are related to dollars with very close degree to that of in terms of cost
```
